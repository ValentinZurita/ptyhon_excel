```python
import pandas as pd
```

# Importing from excel

```python
pd.read_excel("course_participants.xlsx")
```

# Creating data frame

```python
data=[["Mark", 55, "Italy", 4.5, "Europe"],
["John", 33, "USA", 6.7, "America"],
["Tim", 41, "USA", 3.9, "America"],
["Jenny", 12, "Germany", 9.0, "Europe"]]
df = pd.DataFrame(data=data,
                  columns=["name", "age", "country", "score", "continent"],
                  index=[1001, 1000, 1002, 1003])
df
```

```python
df.info()
```

# Index

```python
# Accessing index in a data frame
df.index
```

```python
# Given a name to the index in a data frame
df.index.name = "user_id"
df
```

```python
# "reset_index" turns the index into a column, replacing the index with the default index.
# This corresponds to the DataFrame from the beginning that we loaded from Excel.
df.reset_index()
```

```python
# "reset_index" turns "user_id" into a regular column and
# "set_index" turns the column "name" into the index
df.reset_index().set_index("name")
```

```python
# This methods leave untouched the original data frame as it shows below
df
```

```python
# To change the index, use the reindex method:
df.reindex([999, 1000, 1001, 1004])
```

```python
# This one proves the point on cell no 13
df
```

```python
# To sort an index, use the sort_index method:
df.sort_index()
```

```python
# To sort the rows by one or more columns, use sort_values:
df.sort_values(['continent', "age"])
```

# Colums

```python
# Show columns
df.columns
```

```python
# Change column name
df.columns.name = "properties"
```

```python
# Rename columns
df.rename(columns={"name": "First Name", "age": "Age"})
```

```python
# Delete colums or rows
df.drop(columns=['name', 'country'], index=[1000, 1003])
```

```python
# Transpose
df.T
```

```python
# Reorder the columns
df.loc[:, ["continent", "country", "name", "age", "score"]]
```

# Selecting Data

```python
# Using scalars for both row and column selection returns a scalar
df.loc[1001, "name"]
```

```python
# Using a scalar on either the row or column selection returns a Series
df.loc[[1001, 1002], "age"]
```

```python
# Selecting multiple rows and columns returns a DataFrame
df.loc[:1002, ["name", "country"]]
```

```python
# Shortcut for selecting columns
df["name"]
```

## Selecting by position

```python
# Using interger location to return a scalar
df.iloc[0,0] # first item
```

```python
# Return a series
df.iloc[[0,2],1]
```

```python
# Returns a data frame
df.iloc[:3, [0,2]]
```

## Selecting by boolean index

```python
# Selecting people who live in USA and are older than 40 years
tf = (df["age"] > 40) & (df["country"] == "USA")
```

```python
tf
```

```python
# Show my selection with loc
df.loc[tf, :]
```

```python
# Filter index
df.loc[df.index > 1001, :]
```

```python
# Filter dataframe from participants from Italy and Germany
df.loc[df["country"].isin(["Italy", "Germany"]), :]
```
